```python
def gethtml(url,header):
    r=requests.get(url,headers=header)
    if r.status_code==200:
        return r.text
    else:
        print('网络连接异常')

def util(user_content,keyword):
    if keyword == '昵称':
        for i in user_content:
            if keyword in i:
                return i[3:]
    elif  keyword == '性别':
        for i in user_content:
            if keyword in i:
                return i[3:]
    elif  keyword == '地区':
        for i in user_content:
            if keyword in i:
                return i[3:]
    elif  keyword == '生日':
        for i in user_content:
            if keyword in i:
                return i[3:]
    elif  keyword == '简介':
        for i in user_content:
            if keyword in i:
                return i[3:]
```

## 2 爬取目标用户下的所有微博信息

```python
time_ = []  # 微博发布时间
comment_creater_id = []  #微博的发布者id
comment_url = []  # 某个微博的详细内容url
text = [] #微博文字内容
comment_num = [] #评论数
zan_num = []  # 点赞数
weiboId = []  # 每个微博的唯一编号， 需要id 构造粉丝评论url  https://m.weibo.cn/comments/hotflow?id=4510598555019129&mid=4510598555019129&max_id_type=0

def spider(a):  # 参数a 表示微博的第几页
    url = "https://m.weibo.cn/api/container/getIndex"  # m. 表示手机登入浏览
    # 请求头要改为 手机或者ipad模式
    headers = {'User-Agent': 'Mozilla/5.0 (iPad; CPU OS 11_0 like Mac OS X) AppleWebKit/604.1.34 (KHTML, like Gecko) Version/11.0 Mobile/15A5341f Safari/604.1'}
    data = {  # 这个data  就是 xhr 中 的查询参数
        "uid": "6175460875",
        "type": "uid",
        "value": "6175460875",
        "containerid": "1076036175460875",
        "page": str(a)
    }
    data_html = requests.get(url = url,params=data,headers=headers).json(encoding ="utf-8")  #获取json数据
    # print(data_html)
    for num in  range(1,len(data_html["data"]["cards"])): # len(data_html["data"]["cards"]) 表示每个分页有多少条数据
        if data_html["data"]["cards"][num]["card_type"] == 9:
            time_.append(data_html["data"]["cards"][num]["mblog"]["created_at"])
            comment_creater_id.append(data['uid'])
            text.append(data_html["data"]["cards"][num]["mblog"]["text"])
            weiboId.append(data_html["data"]["cards"][num]["mblog"]["id"])
            comment_url.append(data_html["data"]["cards"][num]["scheme"])
            zan_num.append(data_html["data"]["cards"][num]["mblog"]["attitudes_count"])
            if data_html["data"]["cards"][num]["mblog"]["comments_count"] == "100万+": #将100万+ 转换为数字
                comment_num.append(1000000)
            else:
                comment_num.append(data_html["data"]["cards"][num]["mblog"]["comments_count"])

def run_spider():
    for i in range(0,2):  # 该处可自行定义数量
        print("--正在抓取第【{}】页--".format(i))
        time.sleep(5)
        spider(i)

# print(list(zip(time_,text,comment_url)))

def save_data():
    V = list(zip(weiboId, time_,comment_creater_id,text,comment_url,zan_num,comment_num))
    df = pd.DataFrame(V,columns=["id","time","comment_creater_id","text","comment_url","zan_num","comments_num"])
    df.to_excel("weibo_spider_content.xlsx")
```

```python
run_spider()
save_data()
```

### 3.2 爬取该用户的粉丝id

```python
def get_userid(url):
    header_list = [
        "Opera/12.0(Windows NT 5.2;U;en)Presto/22.9.168 Version/12.00",
        "Opera/12.0(Windows NT 5.1;U;en)Presto/22.9.168 Version/12.00",
        "Mozilla/5.0 (Windows NT 5.1) Gecko/20100101 Firefox/14.0 Opera/12.0",
        "Opera/9.80 (Windows NT 6.1; WOW64; U; pt) Presto/2.10.229 Version/11.62",
        "Opera/9.80 (Windows NT 6.0; U; pl) Presto/2.10.229 Version/11.62",
    ]
    header = {
        'user-agent': random.choice(header_list)
    }

    pat = 'since_id=(.*)'
    with open('user_id.txt', 'w') as f:
        for page in tqdm_notebook(range(1, 10), desc='进度条:'):  #从10页好友中去筛选
            try:
                print(url)
                r = requests.get(url, headers=header)
                all_user = r.json()['data']['cards'][0]['card_group']
                since_id = r.json()['data']['cardlistInfo']['since_id']
                for user in all_user:
                    fans = user['user']['followers_count']
                    if int(fans) > 10:   #当粉丝的粉丝数大于10时，纪录该粉丝的id值，该操作为了避免“机器人”粉丝
                        f.write(str(user.get('user')['id'])+'\n')
                url = re.sub(pat, 'since_id='+str(since_id), url)
                time.sleep(random.randint(1, 2))
            except Exception as e:
                print(e)
```

## 4 数据处理与分析

### 4.1 数据预处理

```python
df = pd.read_excel("num_{}_fans_info.xlsx".format(count))
df
```

```python
df = df[df.birth.notnull()].reset_index(drop=True)   # 过滤birth为空的粉丝
df.birth = df.birth.apply(lambda x: x if x[0]>='0' and x[0]<= '9'  else np.nan)  #去掉birth值为中文情况，如“金牛座”
df.birth = df.birth.apply(lambda x: x if len(str(x))> 5 else np.nan)  #去掉birth值长度小于6情况情况，如“01-01”
df = df[df.birth.notnull()].reset_index(drop=True)   # 过滤birth为空的粉丝
df['age'] = df.birth.apply(lambda x:2022-int(x.split('-')[0]))  #根据birth计算age
df = df[df.location.notnull()].reset_index(drop=True)    # 过滤location为空的粉丝
df['province'] = df.location.apply(lambda x:x.split(' ')[0])
df = df[['user_id','user_name','gender','location','province','birth','age','introduction']]
df = df[df.gender.notnull()].reset_index(drop=True)    # 过滤gender为空的粉丝
df
```
