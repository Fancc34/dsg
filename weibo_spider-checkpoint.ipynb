```python
time_ = []  # å¾®åšå‘å¸ƒæ—¶é—´
comment_creater_id = []  #å¾®åšçš„å‘å¸ƒè€…id
comment_url = []  # æŸä¸ªå¾®åšçš„è¯¦ç»†å†…å®¹url
text = [] #å¾®åšæ–‡å­—å†…å®¹
comment_num = [] #è¯„è®ºæ•°
zan_num = []  # ç‚¹èµæ•°
weiboId = []  # æ¯ä¸ªå¾®åšçš„å”¯ä¸€ç¼–å·ï¼Œ éœ€è¦id æ„é€ ç²‰ä¸è¯„è®ºurl  https://m.weibo.cn/comments/hotflow?id=4510598555019129&mid=4510598555019129&max_id_type=0

def spider(a):  # å‚æ•°a è¡¨ç¤ºå¾®åšçš„ç¬¬å‡ é¡µ
    url = "https://m.weibo.cn/api/container/getIndex"  # m. è¡¨ç¤ºæ‰‹æœºç™»å…¥æµè§ˆ
    # è¯·æ±‚å¤´è¦æ”¹ä¸º æ‰‹æœºæˆ–è€…ipadæ¨¡å¼
    headers = {'User-Agent': 'Mozilla/5.0 (iPad; CPU OS 11_0 like Mac OS X) AppleWebKit/604.1.34 (KHTML, like Gecko) Version/11.0 Mobile/15A5341f Safari/604.1'}
    data = {  # è¿™ä¸ªdata  å°±æ˜¯ xhr ä¸­ çš„æŸ¥è¯¢å‚æ•°
        "uid": "6175460875",
        "type": "uid",
        "value": "6175460875",
        "containerid": "1076036175460875",
        "page": str(a)
    }
    data_html = requests.get(url = url,params=data,headers=headers).json(encoding ="utf-8")  #è·å–jsonæ•°æ®
    # print(data_html)
    for num in  range(1,len(data_html["data"]["cards"])): # len(data_html["data"]["cards"]) è¡¨ç¤ºæ¯ä¸ªåˆ†é¡µæœ‰å¤šå°‘æ¡æ•°æ®
        if data_html["data"]["cards"][num]["card_type"] == 9:
            time_.append(data_html["data"]["cards"][num]["mblog"]["created_at"])
            comment_creater_id.append(data['uid'])
            text.append(data_html["data"]["cards"][num]["mblog"]["text"])
            weiboId.append(data_html["data"]["cards"][num]["mblog"]["id"])
            comment_url.append(data_html["data"]["cards"][num]["scheme"])
            zan_num.append(data_html["data"]["cards"][num]["mblog"]["attitudes_count"])
            if data_html["data"]["cards"][num]["mblog"]["comments_count"] == "100ä¸‡+": #å°†100ä¸‡+ è½¬æ¢ä¸ºæ•°å­—
                comment_num.append(1000000)
            else:
                comment_num.append(data_html["data"]["cards"][num]["mblog"]["comments_count"])

def run_spider():
    for i in range(0,2):  # è¯¥å¤„å¯è‡ªè¡Œå®šä¹‰æ•°é‡
        print("--æ­£åœ¨æŠ“å–ç¬¬ã€{}ã€‘é¡µ--".format(i))
        time.sleep(5)
        spider(i)

# print(list(zip(time_,text,comment_url)))

def save_data():
    V = list(zip(weiboId, time_,comment_creater_id,text,comment_url,zan_num,comment_num))
    df = pd.DataFrame(V,columns=["id","time","comment_creater_id","text","comment_url","zan_num","comments_num"])
    df.to_excel("weibo_spider_content.xlsx")
```

```python
def get_userid(url):
    header_list = [
        "Opera/12.0(Windows NT 5.2;U;en)Presto/22.9.168 Version/12.00",
        "Opera/12.0(Windows NT 5.1;U;en)Presto/22.9.168 Version/12.00",
        "Mozilla/5.0 (Windows NT 5.1) Gecko/20100101 Firefox/14.0 Opera/12.0",
        "Opera/9.80 (Windows NT 6.1; WOW64; U; pt) Presto/2.10.229 Version/11.62",
        "Opera/9.80 (Windows NT 6.0; U; pl) Presto/2.10.229 Version/11.62",
    ]
    header = {
        'user-agent': random.choice(header_list)
    }

    pat = 'since_id=(.*)'
    with open('user_id.txt', 'w') as f:
        for page in tqdm_notebook(range(1, 10), desc='è¿›åº¦æ¡:'):  #ä»10é¡µå¥½å‹ä¸­å»ç­›é€‰
            try:
                print(url)
                r = requests.get(url, headers=header)
                all_user = r.json()['data']['cards'][0]['card_group']
                since_id = r.json()['data']['cardlistInfo']['since_id']
                for user in all_user:
                    fans = user['user']['followers_count']
                    if int(fans) > 10:   #å½“ç²‰ä¸çš„ç²‰ä¸æ•°å¤§äº10æ—¶ï¼Œçºªå½•è¯¥ç²‰ä¸çš„idå€¼ï¼Œè¯¥æ“ä½œä¸ºäº†é¿å…â€œæœºå™¨äººâ€ç²‰ä¸
                        f.write(str(user.get('user')['id'])+'\n')
                url = re.sub(pat, 'since_id='+str(since_id), url)
                time.sleep(random.randint(1, 2))
            except Exception as e:
                print(e)
```
